```python
df["points"] = df["made_free_throws"] + 3*df["made_three_point_field_goals"] + 2*(df["made_field_goals"]-df["made_three_point_field_goals"])
df["rebounds"] = df["offensive_rebounds"] + df["defensive_rebounds"]
```

```python
df[['team']] = df[['team']].fillna('Team.WASHINGTON_BULLETS')
```

```python
def process_team_name(name):
    return name.split('.')[1]
```

```python
df_clean = df[['name','date','team','points','assists','rebounds','steals','blocks','turnovers','personal_fouls',
               'made_three_point_field_goals','attempted_three_point_field_goals']]
df_season9596 = df_clean[(df_clean["date"] >= '2018-10-16') & (df_clean["date"] <= '2019-04-10')]
```

```python
team_dates = {}
for team, df_team in df_season9596.groupby('team', as_index=False):
    aux = df_team.groupby('date', as_index=False).count()
    aux = aux.sort_values(by='date')
    dates = aux["date"]
    aux["rank"] = aux["date"].rank()
    aux = aux[['rank']]
    aux.index = dates
    team_dates[team] = aux.to_dict('index')
```

```python
team_dates
```

```python
def get_game_number(row):
    return team_dates[row['team']][row['date']]['rank']
```

```python
for name, df_name in df_season9596.groupby('name', as_index=False):
    df_name["game_id"] = df_name.apply(lambda row: get_game_number(row), axis=1)
#     pos = players[name.replace(' ','_')]['position']
#     if pos == "": print(name + " doesnt have position")
#     df_name["position"] = pos
    df_name = df_name.sort_values(by='date')
    df_name.to_csv('nba_players_1819/'+name.replace(' ','_'), sep='|', index=False)
```

```python
def process_name(name):
    if ',' in name:
        proc_name = ''
        comma_separated = name.split(',')
        post_comma = comma_separated[1].split(' ')
        pre_comma  = comma_separated[0].split(' ')
        for token in post_comma:
            proc_name += (token + ' ')
        for token in pre_comma:
            proc_name += token + ' '
        return proc_name[1:len(proc_name)-1]
    else:
        return name
```
