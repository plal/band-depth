```python
data_raw = pd.read_csv("AirQualityUCI.csv",sep=";")
columns = ['Date', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)',
           'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)',
           'PT08.S5(O3)', 'T', 'RH', 'AH']
data_raw = data_raw[columns]
data_raw = data_raw.iloc[:9357,:]
data_raw
```

```python
data_selected = data_raw[["Date", "Time", "CO(GT)"]]
data_selected
```

```python
grouped_df = data_selected.groupby( ["Date", "Time"] )
```

```python
for key, item in grouped_df:
    print(grouped_df.get_group(key), "\n\n")
```
